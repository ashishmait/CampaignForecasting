```python
from ipywidgets import widgets
from IPython.display import display

from __future__ import print_function
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
import requests
import json  
import pandas as pd
import numpy as np
from sklearn import neighbors
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.metrics import fbeta_score, make_scorer
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
from sklearn import neighbors
from sklearn.preprocessing import StandardScaler
from ipywidgets import *
%matplotlib inline

df = pd.read_csv('CS_DataComplete_19Apr.csv')

feature_X_All = ['RewardID','TravelProductID' ,'IsPublic','RewardType','RewardEarnLimitCnt','PurchaseAmtMin','PurchaseAmtMax','RewardAmt','RewardPct','hotelCampaign',
                 'days','Include','Exclude','GPG','Redemption']

feature_X = ['RewardID','TravelProductID' ,'IsPublic','RewardType','RewardEarnLimitCnt','PurchaseAmtMin','PurchaseAmtMax','RewardAmt','RewardPct','hotelCampaign',
             'days','Include','Exclude','GPG']

feature_y = ['Redemption']

X_ALL = df[feature_X_All]
X = df[feature_X]
y = df[feature_y]


transactionsText = widgets.Text(description="Transactions ",width=300)
display(transactionsText)

tpidText = widgets.Text(description="Tpid",width=300)
display(tpidText)

similiarCampaignsButton = widgets.Button(description="Get Similiar Campaigns", width=10)
display(similiarCampaignsButton)

similiarCampaigns = []



def get_similiarCampaigns(sender):    

    NEW_X_ALL = X_ALL[(X_ALL.TravelProductID == int(tpidText.value))]
    X_New = NEW_X_ALL[feature_X]
    y_New = NEW_X_ALL[feature_y]
    
    y_actual_value = np.array(y_New)
    
    
    for k in range(0,X_New.shape[0]-1):
        actualValue = y_actual_value[k:k+1]
        expectedValue = int(transactionsText.value)
    
        percentageVariation = abs(float((expectedValue - actualValue[0]))/float(actualValue[0]))*100
    
        if(percentageVariation <= 10):
            print("parameters :" + repr(X_New[k:k+1]) + "Redemption Value : " + repr(actualValue[0]))
            similiarCampaigns.append(X_New[k:k+1])
    
    display(campaignSelectionText)
            

similiarCampaignsButton.on_click(get_similiarCampaigns)

campaignSelectionText = widgets.Text(description="Select Campaign", width=100)

def fill_campaign_details(sender):
    index = int(campaignSelectionText.value)
    campaignDetail = np.array(similiarCampaigns[index:index+1][0])   
    print(campaignDetail)
    
    display(isPublicText)
    isPublicValue = np.uint32(campaignDetail[0][2]).item()
    isPublicText.value = str(isPublicValue)
    
    display(rewardTypeText)
    rewardTypeValue = np.uint32(campaignDetail[0][3]).item()
    rewardTypeText.value = str(rewardTypeValue)
    
    display(rewardEarnLimitText)
    rewardEarnLimitValue = np.uint32(campaignDetail[0][4]).item()
    rewardEarnLimitText.value = str(rewardEarnLimitValue)
    
    display(purchaseAmtMinText)
    purchaseAmtMinValue = np.uint32(campaignDetail[0][5]).item()
    purchaseAmtMinText.value = str(purchaseAmtMinValue)
    
    display(purchaseAmtMaxText)
    purchaseAmtMaxValue = np.uint32(campaignDetail[0][6]).item()
    purchaseAmtMaxText.value = str(purchaseAmtMaxValue)
    
    display(rewardAmntText)
    rewardAmntTextValue = np.float64(campaignDetail[0][7]).item()
    rewardAmntText.value = str(rewardAmntTextValue)
    
    display(rewardPctText)
    rewardPctValue = np.float64(campaignDetail[0][8]).item()
    rewardPctText.value = str(rewardPctValue)
    
    display(ishotelCampaignText)
    ishotelCampaignValue = np.uint32(campaignDetail[0][9]).item()
    ishotelCampaignText.value = str(ishotelCampaignValue)
    
    display(daysText)
    daysTextValue = np.uint32(campaignDetail[0][10]).item()
    daysText.value = str(daysTextValue)
    
    display(hotelIncludeText)
    hotelIncludeValue = np.uint32(campaignDetail[0][11]).item()
    hotelIncludeText.value = str(hotelIncludeValue)
    
    display(hotelExcludeText)
    hotelExcludeValue = np.uint32(campaignDetail[0][12]).item()
    hotelExcludeText.value = str(hotelExcludeValue)
    
    display(gpgText)
    gpgValue = np.uint32(campaignDetail[0][13]).item()
    gpgText.value = str(gpgValue)   
    
    display(redemptionButton)
    display(suggestionsButton)
    
    display(redemptionText)
    
campaignSelectionText.on_submit(fill_campaign_details)

isPublicText = widgets.Text(description="Is Public",width=300)
rewardTypeText = widgets.Text(description="Reward Type",width=300)
rewardEarnLimitText = widgets.Text(description="Reward Earn Limit",width=300)
purchaseAmtMinText = widgets.Text(description="Purchase Amount Min",width=300)
purchaseAmtMaxText = widgets.Text(description="Purchase Amount Max",width=300)
rewardAmntText = widgets.Text(description="Reward Amount",width=300)
rewardPctText = widgets.Text(description="Reward Percentage",width=300)
ishotelCampaignText = widgets.Text(description="Is Hotel Campaign",width=300)
daysText = widgets.Text(description="Days",width=300)
hotelIncludeText = widgets.Text(description="Hotels Included",width=300)
hotelExcludeText = widgets.Text(description="Hotels Excluded",width=300)
gpgText = widgets.Text(description="Gpg",width=300)


redemptionButton = widgets.Button(description="Get Redemption", width=10)
suggestionsButton = widgets.Button(description="Get Suggestions", width=10)

redemptionText = widgets.Text(description="Redemption",width=300)

def get_redemption(sender):
#     print("Button Clicked")
    global jsonRquestObject 
    jsonRquestObject = "{";
    jsonRquestObject = (jsonRquestObject + "\"TPID\":" + tpidText.value + "," 
    + "\"IsPublic\":" + isPublicText.value + "," 
    + "\"RewardType\":" + rewardTypeText.value + ","
    + "\"RewardEarnLimitCnt\":" + rewardEarnLimitText.value + "," 
    + "\"PurchaseAmtMin\":" + purchaseAmtMinText.value + "," 
    + "\"PurchaseAmtMax\":" + purchaseAmtMaxText.value + ","
    + "\"RewardAmt\":" + rewardAmntText.value + ","
    + "\"RewardPct\":" + rewardPctText.value + ","
    + "\"hotelCampaign\":" + ishotelCampaignText.value + ","
    + "\"days\":" + daysText.value + ","
    + "\"Include\":" + hotelIncludeText.value + ","
    + "\"Exclude\":" + hotelExcludeText.value + ","
    + "\"GPG\":" + gpgText.value + "}")    
    
    b = json.loads(jsonRquestObject)
    
    resp = requests.post('http://localhost:5000/campaign/api/v1.0/predictRedemption', json=b)
    print('Created task. ID: {}'.format(resp.json()["redemptionValue"]))
    redemptionVal = resp.json()["redemptionValue"]

    redemptionText.value = str(redemptionVal[0])
    
def get_suggestions(sender):
    global jsonRquestObject 
    jsonRquestObject = "{";
    jsonRquestObject = (jsonRquestObject + "\"TPID\":" + tpidText.value + "," 
    + "\"IsPublic\":" + isPublicText.value + "," 
    + "\"RewardType\":" + rewardTypeText.value + ","
    + "\"RewardEarnLimitCnt\":" + rewardEarnLimitText.value + "," 
    + "\"PurchaseAmtMin\":" + purchaseAmtMinText.value + "," 
    + "\"PurchaseAmtMax\":" + purchaseAmtMaxText.value + ","
    + "\"RewardAmt\":" + rewardAmntText.value + ","
    + "\"RewardPct\":" + rewardPctText.value + ","
    + "\"hotelCampaign\":" + ishotelCampaignText.value + ","
    + "\"days\":" + daysText.value + ","
    + "\"Include\":" + hotelIncludeText.value + ","
    + "\"Exclude\":" + hotelExcludeText.value + ","
    + "\"GPG\":" + gpgText.value + "}")    
    
    print(jsonRquestObject)
    b = json.loads(jsonRquestObject)
    
    resp = requests.post('http://localhost:5000/campaign/api/v1.0/suggestions', json=b)
    #print('Created task. ID: {}'.format(resp.json()["suggestions"]))
    suggestionsVal = resp.json()["suggestions"]

    print(suggestionsVal)
redemptionButton.on_click(get_redemption)
suggestionsButton.on_click(get_suggestions)
```
